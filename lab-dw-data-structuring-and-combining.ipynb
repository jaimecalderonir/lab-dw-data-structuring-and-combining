In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:

But this time, we got new data, which can be found in the following 2 CSV files located at the links below.

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv

Note that you'll need to clean and format the new data.

Observation:
- One option is to first combine the three datasets and then apply the cleaning function to the new combined dataset
- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
# Your code goes here
import pandas as pd
```

```python
df1 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv')
df2 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv')
df3 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv')
```

```python
combined_df = pd.concat([df1, df2, df3], ignore_index=True)


def clean_data(df):
   
    df.columns = df.columns.str.lower().str.replace(' ', '_')
    

    df = df.drop_duplicates()
    
    
    df = df.fillna({'customer_gender': 'Unknown', 'policy_type': 'Unknown', 'total_claim_amount': 0})
    
   
    df['total_claim_amount'] = pd.to_numeric(df['total_claim_amount'], errors='coerce')
    
    return df


cleaned_df = clean_data(combined_df)


cleaned_df.to_csv('cleaned_combined_data.csv', index=False)

```

# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
# Your code goes here
url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv'
df = pd.read_csv(url)
pivot_total_revenue = df.pivot_table(values='total_claim_amount', index='sales_channel', aggfunc='sum')
pivot_total_revenue = pivot_total_revenue.round(2).sort_values(by='total_claim_amount', ascending=False)
print("Total Revenue by Sales Channel:\n", pivot_total_revenue)
pivot_avg_clv = df.pivot_table(values='customer_lifetime_value', index='gender', columns='education', aggfunc='mean')
pivot_avg_clv = pivot_avg_clv.round(2)
print("\nAverage Customer Lifetime Value by Gender and Education Level:\n", pivot_avg_clv)
```

Web channel generates the highest total revenue, indicating a strong online presence and effectiveness in digital marketing strategies.
Branch sales come in second, which shows that physical locations still play a significant role in revenue generation.
Call center is in third place, suggesting that phone-based sales and support are still relevant but less dominant compared to web and branch.
Mail has the lowest revenue, indicating that traditional mail marketing might be the least effective in terms of revenue generation.

Females with a Master's degree have the highest average customer lifetime value, indicating that they might be more loyal or have higher spending power.
Males with a Doctorate also have high lifetime values, suggesting similar trends in loyalty and spending power.
High School education tends to have lower average customer lifetime values for both genders, possibly indicating lower income levels or spending capacity.
Bachelors and College degrees show similar trends, with slight variations between genders.
